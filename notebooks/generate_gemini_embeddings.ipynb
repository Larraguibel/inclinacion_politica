```python
import os
from google import genai
import pandas as pd
import json

from utils.embedding import embed_texts_in_batches

os.environ["GOOGLE_API_KEY"] = ""
client = genai.Client(api_key=os.environ["GOOGLE_API_KEY"])
```

```python
cajcodes = pd.read_csv('../data/hf_datasets/cajcodes.csv')
tweets = pd.read_csv('../data/hf_datasets/political-leaning-tweets-filtered.csv')
gen_news = pd.read_json("../data/hf_datasets/generated_news_political_leaning.json")
```

```python
BATCH_SIZE = 100   # puedes ajustar (500–1000 es seguro)
MODEL_NAME = "text-embedding-004"
```

```python
print("Generando embeddings...")
embeddings_cajcodes = embed_texts_in_batches(cajcodes.text.tolist(), model=MODEL_NAME, batch_size=BATCH_SIZE)

file_name = '../data/embeddings/cajcodes-embeddings.parquet'
df = pd.DataFrame({
    "text": cajcodes.text,
    "label": cajcodes.label,
    "embedding": [json.dumps(e.tolist()) for e in embeddings_cajcodes]
    })
df.to_parquet(file_name, compression="snappy")
print('Guardados embeddings cajcodes\n')
```

```python
print("Generando embeddings...")
embeddings_tweets = embed_texts_in_batches(tweets.text.tolist(), model=MODEL_NAME, batch_size=BATCH_SIZE)

file_name = '../data/embeddings/tweets-embeddings.parquet'
df = pd.DataFrame({
    "text": tweets.text,
    "label": tweets.label,
    "embedding": [json.dumps(e.tolist()) for e in embeddings_tweets]
    })
df.to_parquet(file_name, compression="snappy")
print('Guardados embeddings tweets\n')
```

```python
print("Generando embeddings...")
embeddings_gen_news = embed_texts_in_batches(gen_news.text.tolist(), model=MODEL_NAME, batch_size=BATCH_SIZE)

file_name = '../data/embeddings/gen_news.parquet'
df = pd.DataFrame({
    "text": gen_news.text,
    "label": gen_news.label,
    "embedding": [json.dumps(e.tolist()) for e in embeddings_gen_news]
    })
df.to_parquet(file_name, compression="snappy")
print('Guardados embeddings Gen News\n')
```
