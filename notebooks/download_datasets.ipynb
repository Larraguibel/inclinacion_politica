### ```Generated_News_Political_Leaning```

````python
ds = load_dataset("FilipT/Generated_News_Political_Leaning")
gnpl = pd.DataFrame(ds['train'])
````

````python
gnpl= gnpl[[ "generation","category", "main_domain"]]
gnpl.drop_duplicates(inplace=True)
gnpl.rename(columns={"generation": "text", "category": "label"}, inplace=True)
label_map = {"Right-wing": "R", "Unbiased": "C", "Left-wing": "L"}
gnpl["label"] = gnpl["label"].map(label_map)
````

````python
gnpl.sample(10, random_state=42)
````

````python
gnpl.to_json("../data/hf_datasets/generated_news_political_leaning.json")
````
